**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
# 先求出1968年~2015年六月的df   movie_data_34
movie_data_34 = movie_data[(movie_data['release_date'].astype('datetime64').dt.month == 6) & (movie_data['release_year'] > 1967)]
movie_data_34.describe()
movie_counts = movie_data_34['release_year'].value_counts()
base_color = sb.color_palette()[0]
plt.figure(figsize=(20,6))
sb.countplot(data = movie_data_34, x = 'release_year',color= base_color);
plt.xticks(rotation = 90)
locs,labels = plt.xticks()
for loc,label in zip(locs,labels):
    count = movie_counts[int(label.get_text())]
    plt.text(loc, count+0.5, str(count), ha = 'center')
```

# 结果分析
由上图可以看出，从1968年到2015年之间6月份的电影数量总体是上升趋势，其中2014年时最多，一共是是48部

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
movie_count_Comedy = movie_data_34[movie_data_34['genres'].str.contains('Comedy',regex=False)][['genres','release_year']]
movie_count_Comedy['genres'] = 'Comedy'
movie_count_Drama = movie_data_34[movie_data_34['genres'].str.contains('Drama',regex=False)][['genres','release_year']]
movie_count_Drama['genres'] = 'Drama'
movie_count = movie_count_Comedy.merge(movie_count_Drama,how ='outer').sort_values(by = 'release_year',axis = 0,ascending = True)
def drawing(x,**kwargs):
    y_means = x.groupby(x).count()
    plt.errorbar(x = y_means.index.astype(str), y = y_means,**kwargs)
g = sb.FacetGrid(data = movie_count,hue='genres', height = 10)
g.map(drawing,'release_year')
g.add_legend()
plt.xticks(rotation = 90);
```

# 结果分析
由上图可以看出1968年~2015年六月期间Comedy类电影的数量和Drama电影的数量整体都成上升趋势
